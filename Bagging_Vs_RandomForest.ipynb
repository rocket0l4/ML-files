#Creating dataset by make_classification

```python
import numpy as np
import pandas as pd
from sklearn.datasets import make_classification
from sklearn.ensemble import BaggingClassifier,RandomForestClassifier
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
```

```python
X,y = make_classification(n_features=5,n_redundant=0,n_informative=5,n_clusters_per_class=1)
```

#Required columns in Dataset

```python
df = pd.DataFrame(X,columns=['col1','col2','col3','col4','col5'])
df['target']  = y
print(df.shape)
df.head()
```

#Making of Bagging Classifier
- max_feature suggest column sampling

# here we clearly see, in bagging the tree level sampling occure(which means heer it is pre-definded of the column sampliing, and we only use that columns only)

```python
bag = BaggingClassifier(max_features=2)
```

```python
bag.fit(df.iloc[:,:5],df.iloc[:,-1])
```

```python
plt.figure(figsize=(12,12))
plot_tree(bag.estimators_[0])
plt.show()
```

```python
rt = RandomForestClassifier(max_features=2)
```

#Making RandomForest

- here we clearly see the node level splitting occure.
- here selection occure randomly.

```python
rt.fit(df.iloc[:,:5],df.iloc[:,-1])
```

```python
plt.figure(figsize=(12,12))
plot_tree(rt.estimators_[4])
plt.show()
```
