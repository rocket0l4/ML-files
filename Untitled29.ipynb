```python
def evaluate(clf,X,y):
  clf.fit(X,y)
  plot_tree(clf)
  plt.show()
  plot_decision_regions(X.values,y.values,clf=clf,legend=2)
  y_pred = clf.predict(X_test)
  print(accuracy_score(y_test,y_pred))
```
