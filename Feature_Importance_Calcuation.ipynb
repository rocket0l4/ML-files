#Creating dataset with 2 columns.

```python
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import make_classification
```

```python
X,y = make_classification(n_samples=15,n_features=2,n_redundant=0,n_classes=2,random_state=0)
```

```python
clf = DecisionTreeClassifier()
```

```python
clf.fit(X,y)
```

```python
from sklearn.tree import plot_tree
```

```python
plot_tree(clf)
```

# As we clearly see there are 2 nodes present which are splitting Node-1 and Node-2, and we need to figure out node importance the using node impurity formula...

- Formula for node importance
- Ni = N_T/N[impurity - ((N_T_r/N_t) x right_impurity)-((N_T_L/N_t) x left_imputity)]

N = Total no. of Rows.
N_t = no of rows of a particular node.
N_t_r = Right Node.
N_t_L = Lef Node.
