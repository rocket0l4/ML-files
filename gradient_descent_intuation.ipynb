```python
from sklearn.datasets import make_regression
import numpy as np
```

```python
X,y = make_regression(n_samples=4,n_features=1,n_informative=1,n_targets=1,noise=80,random_state=13)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X,y)
```

```python
# Applying OLS by using sklearn
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
reg.fit(X,y)
```

```python
# m
reg.coef_
```

```python
# b
reg.intercept_
```

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color = 'red')
```

```python
# Applying the Gradient Descent assuming slop is constant m = 78.35
# Let assume the starting value for intercept b = 0
# Now calculating the value for y_pred
y_pred = ((78.35 * X)+0).reshape(4)
```

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color = 'red',label = 'OLS')
plt.plot(X,y_pred,color = 'green',label = 'b = 0')
plt.legend()
plt.show()
```

```python
m = 78.35
b  = 0

loss_slop = -2 * np.sum(y - (m*X.ravel())-b)
loss_slop
```

```python
# Let's take learning rate = 0.1
# learning rate * loss slop is equal to step_size
lr = 0.1

step_size = loss_slop*lr
step_size
```

```python
# Now calculating the new intercept, i.e b
b = b - step_size
b
```

```python
# Now calculating for the new b value
y_pred1 = ((78.35 * X)+b).reshape(4)

plt.scatter(X,y)
plt.plot(X,reg.predict(X),color = 'red',label = 'OLS')
plt.plot(X,y_pred1,color = 'black',label ='b = {}'.format(b))
plt.plot(X,y_pred,color = 'blue',label = 'b = 0')
plt.legend()
plt.show()
```

```python
#iteration 2
loss_slop = -2 * np.sum(y - (m*X.ravel())-b)
loss_slop
```

```python
step_size = loss_slop*lr
step_size
```

```python
b = b - step_size
b
```

```python
y_pred2 = ((78.35*X) +b).reshape(4)

plt.scatter(X,y)
plt.plot(X,reg.predict(X),color = 'red',label = 'OLS')
plt.plot(X,y_pred2,color = 'grey',label = 'b = {}'.format(b))
plt.plot(X,y_pred1,color = 'black',label ='b = {}'.format(b))
plt.plot(X,y_pred,color = 'blue',label = 'b = 0')
plt.legend()
plt.show()
```

```python
# iteration 3
loss_slop = -2 * np.sum(y - (m*X.ravel())-b)
loss_slop
```

```python
step_size = loss_slop*lr
step_size
```

```python
b = b- step_size
b
```

```python
y_pred3 = ((78.35 * X)+b ).reshape(4)

plt.scatter(X,y)
plt.plot(X,reg.predict(X),color = 'red',label = 'OLS')
plt.plot(X,y_pred3,color ='orange',label = 'b {}'.format(b))
plt.plot(X,y_pred2,color ='grey',label = 'b {}'.format(b))
plt.plot(X,y_pred1,color = 'black',label ='b = {}'.format(b))
plt.plot(X,y_pred,color = 'blue',label = 'b = 0')
plt.legend()
plt.show()
```

```python
b = -100
m = 78.35
lr = 0.01

epochs  = 50000

for i in range(epochs):
  loss_slop = -2 * np.sum(y-(m*X.ravel())-b)
  b = b - (lr * loss_slop)
  y_pred  = m*X+b

  plt.plot(X,y_pred)

plt.scatter(X,y)
```
