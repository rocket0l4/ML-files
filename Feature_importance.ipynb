#Importing the Mnist dataset...

```python
from tensorflow.keras.datasets import mnist
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python

(X_train,y_train),(X_test,y_test) = mnist.load_data()
```

#Creating DataFrame

```python
df = pd.DataFrame(X_train.reshape(X_train.shape[0],-1))
```

```python
df
```

```python
X = df.iloc[:,1:]
y = df.iloc[:,0]
```

```python
reshaped_data = np.append(X.iloc[5].values,0).reshape(28,28)
```

```python
sns.heatmap(reshaped_data)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier()
```

```python
rf.fit(X,y)
```

#Giving the importance of every Columns

```python
rf.feature_importances_
```

```python
# reshaped_data = np.append(X.iloc[5].values,0).reshape(28,28)
```

```python
dim = int(np.sqrt(rf.feature_importances_.shape[0]))
```

```python
reshaped_importances = rf.feature_importances_[:dim*dim].reshape(dim,dim)
```

#This heatmap is trying to show that this pink reason is important, as we clearly see the are only 0 present in the dataset(may be due to some bad dataset ) but due to this it is showing that the all the pixcles/points are important..

```python
sns.heatmap(reshaped_importances)
plt.show()
```
