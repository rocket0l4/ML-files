#Loading mnist dataset

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('/content/drive/MyDrive/emnist.csv')
```

```python
df
```

#Train test split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(df.iloc[:,1:],df.iloc[:,0],test_size=0.2,random_state =2)
```

#Loading models

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
```

```python
m1 = LogisticRegression()
m2 = DecisionTreeClassifier()
```

```python
m1.fit(X_train,y_train)
m2.fit(X_train,y_train)
```

```python
y_pred1 = m1.predict(X_test)
y_pred2 = m2.predict(X_test)
```

#Loading metrics
- accurancy socre check.

```python
from sklearn.metrics import accuracy_score,confusion_matrix
print("Accuracy of Logistic Regression",accuracy_score(y_test,y_pred1))
print("Accuracy of Decision Tree",accuracy_score(y_test,y_pred2))
```

#Printting consusion metrics for logistic Regression.

```python
print("Logistic regression confustion Matrix\n")
pd.DataFrame(confusion_matrix(y_test,y_pred1),columns=list(range(0,10)))
```

#Printting consusion metrics for Decision Tree.

```python
print("Decision Tree Confusion Matrix\n")
pd.DataFrame(confusion_matrix(y_test,y_pred2),columns=list(range(0,10)))
```

#Checking socres by different-different metrics.

```python
from sklearn.metrics import precision_score,recall_score,f1_score
```

```python
print("Precision-score :",precision_score(y_test,y_pred1,average='weighted')) # average can be  = 'macro',None,'weighted'
print("Recall-score :",recall_score(y_test,y_pred1,average='weighted'))
print("F1-score :",f1_score(y_test,y_pred1,average='weighted'))
```

#Sklearn gives single way to calculate in one step "classification_report"

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred1))
```
